```python
lyrics_dictionary = []
missed_songs = []
counter = 0


# loop through all 1,000 song dictionaires
for song, row in df.iterrows():
    counter = counter + 1
    print(counter, "Search count")
    #Trying to search for song
    try:
      lg_song = genius.search_song(title=row[1], artist=row[2])
    #if not, song added to alternative dictionary
    except Exception:
        error_dict = {
            "title": row[2],
            "artist": row[1],
            "id": row[5]
        }
        missed_songs.append(error_dict)
        continue
    #If there is a successful API call and the song is found, recreate a new dictionary similar to the spotify api, but include song lyrics down the bottom
    if lg_song:
        lyrics = {
            "title": row[0],
            "artist": row[1],
            "id": row[5],
            "year_released": row['YEAR'],
            "danceability": row['danceability'],
            "energy": row['energy'],
            "key": row['key'],
            "loudness": row['loudness'],
            "mode": row['mode'],
            "speechiness": row['speechiness'],
            "acousticness": row['acousticness'],
            "instrumentalness": row['instrumentalness'],
            "liveness": row['liveness'],
            "valence": row['valence'],
            "tempo": row['tempo'],
            "duration_ms": row['duration_ms'],
            #"date": song_data["date"],
            #song lyrics included in this dictionary
            "lyrics": lg_song.lyrics
        }
        lyrics_dictionary.append(lyrics)
```
